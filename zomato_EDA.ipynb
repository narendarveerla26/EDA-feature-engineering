```python
agg=data.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
agg.reset_index().head()##to create dataframe
```

```python
ratings=agg.reset_index().rename(columns={0:'Rating count'})##rename column name
```

```python
ratings
```

```python
###observation: 
4.5 to 4.9 .....excellent('Dark green')
3.5  to 3.9 ....Good('yellow')
4.0 to 4.4.....very Good('Green')
2.5 to 3.4.....average('orange')
1.8 to 2.4........poor('Red')
0.....no ratings('white')
```

```python

import matplotlib
matplotlib.rcParams['figure.figsize']=(15,10)
sns.barplot(x='Aggregate rating',y='Rating count',data=ratings)
```

```python
matplotlib.rcParams['figure.figsize']=(20,10)
sns.barplot(x='Aggregate rating',y='Rating count',hue='Rating color',data=ratings,palette=['Blue','Red','Orange','Yellow','Green','Green'])
```

```python
#observations :
#No rating counts are very high
#max no of ratings are b/w 2.8 to 4.0
```

```python
sns.countplot(x='Rating color',data=ratings,palette=['Blue','Red','Orange','Yellow','Green','Green'])
```

```python
##this is a not rating count ,thisis frequency of ratings
```
